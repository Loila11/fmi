```python
# 1. download file
import urllib

link = "https://www.gutenberg.org/"
url_response = urllib.request.urlopen(link)
url_contents = url_response.read()
url_response.close()

print(url_contents)
```

```python
# 2. remove header
title_start = url_contents.find(b'<h1')
url_contents = url_contents[title_start:]
url_contents = url_contents.decode('utf-8')

print(url_contents)
```

```python
# 3. Print the number of sentences in the text.
import nltk.data

sentences = nltk.tokenize.sent_tokenize(url_contents)

print(len(sentences))

# Print the average length (number of words) of a sentence.
print(sum([len(sentence) for sentence in sentences]) / len(sentences))
```

```python
# 4. Find the collocations in the text (bigram and trigram)
from nltk.collocations import *

# bigram
bigram_measures = nltk.collocations.BigramAssocMeasures()

finder = BigramCollocationFinder.from_words(nltk.wordpunct_tokenize(url_contents))
scored = finder.score_ngrams(bigram_measures.raw_freq)
sorted(bigram for bigram, score in scored)

# trigram
trigram_measures = nltk.collocations.TrigramAssocMeasures()

finder = TrigramCollocationFinder.from_words(nltk.wordpunct_tokenize(url_contents))
scored = finder.score_ngrams(bigram_measures.raw_freq)
sorted(bigram for bigram, score in scored)
```

```python
# 5. Create a list of all the words (in lower case) from the text, without the punctuation.
import string

words = [word.lower() for word in nltk.wordpunct_tokenize(url_contents) if word[0] not in string.punctuation]
words
```

```python
# 6. Print the first N most frequent words (alphanumeric strings) together with their number of appearances.
import operator
words_dict = {}

for word in words:
    if not word in words_dict:
        words_dict[word] = 0
    
    words_dict[word] += 1

N = 10
sorted(words_dict.items(), key=operator.itemgetter(1), reverse=True)[:N]
```

```python
# 7. Remove stopwords and assign the result to variable lws
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

lws = [word for word in words if not word in stop_words]
lws
```

```python
# 8. Apply stemming (Porter) on the list of words (lws). 
# Print the first 200 words. 
ps = nltk.PorterStemmer()

[ps.stem(word) for word in lws[:200]]

# Do you see any words that don't appear in the dictionary?
# Yes
```

```python
# 9. Print a table of three columns (of size N, where N is the maximum length for the words in the text) [...]
ls=nltk.LancasterStemmer()
snb=nltk.SnowballStemmer("english")

NW = 500
words_set = list(set(words))[:NW]

print('Porter    | Lancaster    | Snowball')
print('-----------------------------------')
print('\n'.join(list(f'{ps.stem(word)}   | {ls.stem(word)}    | {snb.stem(word)}' for word in words_set
                if ps.stem(word) != ls.stem(word) or ps.stem(word) != snb.stem(word))))
```

```python
# 10. Print a table of two columns that will compare the results of stemming and lemmatization [...]
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()

NW = 500
words_set = list(set(words))[:NW]

print('Snowball | WordNetLemmatizer')
print('----------------------------')
print('\n'.join(list(f'{snb.stem(word)} | {lem.lemmatize(word)}' for word in words_set
                if snb.stem(word) != lem.lemmatize(word))))
```

```python
# 11. Print the first N most frequent lemmas (after the removal of stopwords) together with their number of appearances.
lemmas_dict = {}

for word in lws:
    lemma = lem.lemmatize(word)
    if not lemma in lemmas_dict:
        lemmas_dict[lemma] = 0
    
    lemmas_dict[lemma] += 1

N = 10
sorted(lemmas_dict.items(), key=operator.itemgetter(1), reverse=True)[:N]
```

```python
# 12. Change all the numbers from lws into words. 
# Print the number of changes, and also the portion of list that contains first N changes (for example N=10).
!pip install num2words
from num2words import num2words

count = 0
N = 10
numeric_lws = []

for word in lws:
    if word.isnumeric():
        numeric_lws.append(num2words(word))
        count += 1
        
        if count == 10:
            print(numeric_lws)
    else:
        numeric_lws.append(word)

print(count)
```

```python
# 13. print the concordance data for the word W
def concordance_data(N, W, l):
    for i in range(len(l)):
        if l[i] == W:
            print(l[max(0, i - int(N/2)):min(len(l), i + int(N/2) + 1)])

concordance_data(3, 'women', lws)
```

```python
# Create a second function that prints windows of texts that contain words only from the phrase containing word W
def phraseW(N, W):
    for sentence in sentences:
        concordance_data(N, W, sentence)

phraseW(3, 'women')
```
