```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = np.zeros((9, 400, 600))
for i in range(9):
    images[i] = np.load('images\car_' + str(i) + '.npy')
```

```python
sum_a = np.sum(images)
print(sum_a)
```

```python
sum_b = np.sum(images, (1, 2))
print(sum_b)
```

```python
print(np.argmax(sum_b))
```

```python
mean_image = np.mean(images, 0)
print(mean_image)
```

```python
plt.imshow(mean_image)
```

```python
dev_std = np.std(images)
print(dev_std)
```

```python
norm_images = (images - mean_image) / dev_std
for i in range(9):
    plt.imshow(norm_images[i])
    plt.show()
```

```python
crop = images[:, 200:301, 280:401]
for i in range(9):
    plt.imshow(crop[i])
    plt.show()
```
